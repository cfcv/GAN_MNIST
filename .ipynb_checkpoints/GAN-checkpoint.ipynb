# Generative Adversarial Network

## Step 1: reading the MNIST dataset 

```python
%matplotlib inline

import pickle as pkl
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data')
```

## Step 2: Building the inputs 

```python
def model_inputs(real_dim, z_dim):
    input_real = tf.placeholder(tf.float32, (None, real_dim), name='input_real')
    input_z = tf.placeholder(tf.float32, (None, z_dim), name='input_z')
    return input_real, input_z
```

## Step 3: Building the Generator Network 

We wil use tf.variable scope to ensure that the trainable variables names of the generator and the discriminator starts with 'generator' or 'discriminator' respectvely  

```python
def Generator(input_z, out_dim, n_units, reuse=False, alpha=0.01):
    with tf.variable_scope('generator', reuse=reuse):
        h1 = tf.layers.dense(input_z, n_units, activation=None)
        h1 = tf.maximum(alpha*h1, h1)
        h2 = tf.layers.dense(h1, n_units*2, activation=None)
        h2 = tf.maximum(alpha*h2, h2)

        logits = tf.layers.dense(h2, out_dim, activation=None)
        out = tf.tanh(logits)
        return out
```

## Step 4: Building the Discriminator Network 

```python
def Discriminator(input_d, n_units, reuse=False, alpha=0.01):
    with tf.variable_scope('discriminator', reuse=reuse):
        h1 = tf.layers.dense(input_d, n_units, activation=None)
        h1 = tf.maximum(alpha*h1, h1)
        h2 = tf.layers.dense(h1, n_units*2, activation=None)
        h2 = tf.maximum(alpha*h2, h2)
        
        logits = tf.layers.dense(h2, 1, activation=None)
        out = tf.sigmoid(logits)
        
        return out, logits
```

## Step 5: Building the Network 

Here i'm defining the hyperparameters and building one generator and two discriminators, one for real data and other for fake data. As we want that the two of then have the same weights we have to share variables so we use reuse=True in the fake data discriminator  

```python
tf.reset_default_graph()

#Hyperparameters
z_size = 100
input_size = 784 #28x28 flatten mnist images
g_hidden_size = 128
d_hidden_size = 128
alpha = 0.01
smooth = 0.1
learning_rate = 0.002


#networks inputs
input_real, input_z = model_inputs(input_size, z_size)

#Generator
g_model = Generator(input_z, input_size, n_units=g_hidden_size, alpha=alpha)

#Discriminator for real data
d_model_real, d_logits_real = Discriminator(input_real, n_units=d_hidden_size, alpha=alpha)

#Discriminator for fake data
d_model_fake, d_logits_fake = Discriminator(g_model, n_units=d_hidden_size, reuse=True, alpha=alpha)
```

#### Discriminator and Generator losses 

The discriminator loss is the sum of the discriminator loss for real and fake data. To improve performance i'm using label smoothing in the real data discriminator labels. For the generator we pass the fake data discriminator loss and the label as ones because we want the generator to generate images that the discriminator predict as true images  

```python
d_loss_real = tf.reduce_mean(
                            tf.nn.sigmoid_cross_entropy_with_logits(logits=d_logits_real,
                                                                    labels=tf.ones_like(d_logits_real)*(1-smooth)))

d_loss_fake = tf.reduce_mean(
                            tf.nn.sigmoid_cross_entropy_with_logits(logits=d_logits_fake,
                                                                    labels=tf.zeros_like(d_logits_real)))
d_loss = d_loss_real + d_loss_fake

g_loss = tf.reduce_mean(
                        tf.nn.sigmoid_cross_entropy_with_logits(logits=d_logits_fake,
                                                                labels=tf.ones_like(d_logits_fake)))
```

#### Optimizers 

As we want to train the discriminator and the generator separately we have to split the trainable variables and create separeted optimizers 

```python
t_vars = tf.trainable_variables()

g_vars = [var for var in t_vars if var.name.startswith('generator')]
d_vars = [var for var in t_vars if var.name.startswith('discriminator')]

d_opt = tf.train.AdamOptimizer(learning_rate).minimize(d_loss, var_list=d_vars)
g_opt = tf.train.AdamOptimizer(learning_rate).minimize(g_loss, var_list=g_vars)
```

## Training

```python
batch_size = 100
epochs = 200
samples = []
losses = []

saver = tf.train.Saver(var_list=g_vars)
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for e in range(epochs):
        for ii in range(mnist.train.num_examples//batch_size):
            batch = mnist.train.next_batch(batch_size)
            
            #First we have to reshape the images to pass trought the disciminator
            #And also rescale because the tahn activation generates samples from a range of -1 to 1 
            #and not 0 to 1
            batch_images = batch[0].reshape((batch_size, 784))
            batch_images = 2*batch_images - 1
            
            #sample random noise for the generator
            batch_z = np.random.uniform(-1, 1, size=(batch_size, z_size))
            
            #run optimizers
            _ = sess.run(d_opt, feed_dict={input_real: batch_images, input_z: batch_z})
            _ = sess.run(g_opt, feed_dict={input_z: batch_z})
            
        #get the losses in the end of the epoch en print
        d_train_loss = sess.run(d_loss, {input_real: batch_images, input_z: batch_z})
        g_train_loss = g_loss.eval({input_z: batch_z})
            
        print("Epoch {}/{}...".format(e+1, epochs),
              "Discriminator Loss: {:.4f}...".format(d_train_loss),
              "Generator Loss: {:.4f}".format(g_train_loss))
            
        losses.append((d_train_loss, g_train_loss))
            
        #sample from generator while training for viewing after
        sample_z = np.random.uniform(-1, 1, size=(16, z_size))
        gen_samples = sess.run(
                            Generator(input_z, input_size, n_units=g_hidden_size, reuse=True, alpha=alpha),
                            feed_dict={input_z: sample_z})
            
        samples.append(gen_samples)
        saver.save(sess, './checkpoints/generator.ckpt')

# Save training generator samples
with open('train_samples.pkl', 'wb') as f:
    pkl.dump(samples, f)
```

```python
fig, ax = plt.subplots()
losses = np.array(losses)
plt.plot(losses.T[0], label='Discriminator')
plt.plot(losses.T[1], label='Generator')
plt.title("Training Losses")
plt.legend()
```

```python
def view_samples(epoch, samples):
    fig, axes = plt.subplots(figsize=(7,7), nrows=4, ncols=4, sharey=True, sharex=True)
    for ax, img in zip(axes.flatten(), samples[epoch]):
        ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)
        im = ax.imshow(img.reshape((28,28)), cmap='Greys_r')
    
    return fig, axes
```
